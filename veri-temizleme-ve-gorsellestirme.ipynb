## Google Play Store EDA & Data Visualization


###### App: Name of the app
###### Category: Category of the app. Some examples are: ART_AND_DESIGN, FINANCE, COMICS, BEAUTY etc.
###### Rating: The current average rating (out of 5) of the app on Google Play
###### Reviews: Number of user reviews given on the app
###### Size: Size of the app in MB (megabytes)
###### Installs: Number of times the app was downloaded from Google Play
###### Type: Whether the app is paid or free
###### Price: Price of the app in US$
###### Last Updated: Date on which the app was last updated on Google Play

```python
import pandas as pd
import seaborn as sns #eksik verileri görselleştireceğiz
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('googleplaystore.csv')
```

```python
#veri setinin ilk satırlarına bakılır
df.head() 
```

```python
# sütun isimlerini görüntülüyoruz
df.columns
```

```python
df.columns = df.columns.str.replace(" ", "_")
```

```python
df.columns
```

```python
#satır-sütun sayısı görüntüleme
df.shape
```

```python
#sütunlarının veri tipini kontrol ediyoruz
df.dtypes
```

```python
#eksik veri analizi
df.isnull().sum()
```

```python
#grafiklerin görünümünü iyileştirme
sns.set_theme()
#grafik boyut ayarlama
sns.set(rc={"figure.dpi":300, "figure.figsize":(12,9)})
```

```python
#eksik verileri görselleştirme
sns.heatmap(df.isnull(), cbar=False)
```

```python
#sütunun medyan değerini bulma
rating_median = df["Rating"].median()
rating_median
```

```python
#rating sütunundaki eksik verileri doldurduk
df["Rating"].fillna(rating_median, inplace=True)
```

```python
#eksik veri silme
df.dropna(inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
#veri setimizdeki sütunlar hakkında özet bilgi
df.info()
```

```python
#review sütununu sayısala çevirme
df["Reviews"].describe()
```

```python
df["Reviews"] = df["Reviews"].astype("int64")
```

```python
df["Reviews"].describe().round
```

```python
print(len(df["Size"].unique()))
df["Size"].unique()
```

```python
#size sütunundeki M ve k 'yı kaldırıyoruz
df["Size"].replace("M", "", regex=True, inplace=True)
df["Size"].replace("k", "", regex=True, inplace=True)
```

```python
print(len(df["Size"].unique()))
df["Size"].unique()
```

```python
size_median =df[df["Size"]!="Varies with device"]["Size"].astype(float).median()
size_median
```

```python
#metin yerine medyan değerini atıyoruz
df["Size"].replace("Varies with device", size_median, inplace=True)
```

```python
#size sütunundaki stringleri kaldırdık ve sayısala çevirdik
df.Size = pd.to_numeric(df.Size)
```

```python
df.Size.head()
```

```python
df.Size.describe().round()
```

```python
df["Installs"].unique()
```

```python
#Installs sütunundaki + değerlerini kaldırıyoruz
df.Installs = df.Installs.apply(lambda x:x.replace("+",""))
df.Installs = df.Installs.apply(lambda x:x.replace(",",""))
df.Installs = df.Installs.apply(lambda x:int(x))
```

```python
df["Installs"].unique()
```

```python
df["Price"].unique()
```

```python
#Price sütunundaki $ ifadesini kaldırıyoruz
df.Price = df.Price.apply(lambda x:x.replace("$", ""))
df.Price = df.Price.apply(lambda x:float(x))
```

```python
df["Price"].unique()
```

```python
len(df["Genres"].unique())
```

```python
df["Genres"].head(10)
```

```python
#yalnızca tür olarak yazdırıyoruz
df["Genres"] = df["Genres"].str.split(";").str[0]
```

```python
len(df["Genres"].unique())
```

```python
df["Genres"].unique()
```

```python
df["Genres"].value_counts()
```

```python
df["Genres"].replace("Music & Audio", "Music", inplace=True)
```

```python
df["Last_Updated"].head()
```

```python
df["Last_Updated"] = pd.to_datetime(df["Last_Updated"])
```

```python
#EDA sonrası veri setine bakıyoruz
df.head()
```
